```python
### all functions 


#### default setting 
###  ROI ###
# HBO
Dorsal_L_o=['S1_D1 hbo','S1_D2 hbo','S2_D1 hbo','S2_D3 hbo','S3_D2 hbo','S3_D3 hbo']
Dorsal_R_o=['S6_D5 hbo','S6_D6 hbo','S7_D5 hbo','S7_D7 hbo','S8_D6 hbo','S8_D7 hbo']
PreFrontal_o=['S3_D4 hbo','S4_D2 hbo','S4_D4 hbo','S4_D5 hbo','S5_D3 hbo','S5_D4 hbo','S5_D6 hbo','S6_D4 hbo']

PreMoter_R_o=['S10_D10 hbo','S9_D9 hbo','S9_D10 hbo','S11_D10 hbo','S11_D12 hbo']
Pri_Moter_R_o=['S10_D11 hbo','S10_D9 hbo','S10_D12 hbo','S12_D11 hbo','S12_D12 hbo']

PreMoter_L_o=['S13_D13 hbo','S13_D14 hbo','S15_D14 hbo','S16_D14 hbo','S16_D16 hbo']
Pri_Moter_L_o=['S14_D13 hbo','S14_D15 hbo','S15_D13 hbo','S15_D15 hbo','S15_D16 hbo']

# HBR
Dorsal_L_r=['S1_D1 hbr','S1_D2 hbr','S2_D1 hbr','S2_D3 hbr','S3_D2 hbr','S3_D3 hbr']
Dorsal_R_r=['S6_D5 hbr','S6_D6 hbr','S7_D5 hbr','S7_D7 hbr','S8_D6 hbr','S8_D7 hbr']
PreFrontal_r=['S3_D4 hbr','S4_D2 hbr','S4_D4 hbr','S4_D5 hbr','S5_D3 hbr','S5_D4 hbr','S5_D6 hbr','S6_D4 hbr']

PreMoter_R_r=['S10_D10 hbr','S9_D9 hbr','S9_D10 hbr','S11_D10 hbr','S11_D12 hbr']
Pri_Moter_R_r=['S10_D11 hbr','S10_D9 hbr','S10_D12 hbr','S12_D11 hbr','S12_D12 hbr']

PreMoter_L_r=['S13_D13 hbr','S13_D14 hbr','S15_D14 hbr','S16_D14 hbr','S16_D16 hbr']
Pri_Moter_L_r=['S14_D13 hbr','S14_D15 hbr','S15_D13 hbr','S15_D15 hbr','S15_D16 hbr']

### read envents 
FileName_B= './Results/Events/events_B.csv'
df_raw_B = pd.read_csv(FileName_B)

FileName_P= './Results/Events/events_P.csv'
df_raw_P = pd.read_csv(FileName_P)

FileName_G= './Results/Events/events_G.csv'
df_raw_G = pd.read_csv(FileName_G)

### fnirs preprocess
def GetRawIntensity(fileName):
    ### read raw snirf 
    snirf_intensity = read_raw_nirx(fname=fileName)
    subject = snirf_intensity.info['subject_info']['his_id']
    return snirf_intensity, subject

def GetRawOd(raw_intensity, plot = False):
    ### deal with density ###
    raw_od = mne.preprocessing.nirs.optical_density(raw_intensity)
    ### detect bad channels
    sci = mne.preprocessing.nirs.scalp_coupling_index(raw_od)
    if plot:
        fig, ax = plt.subplots()
        ax.hist(sci)
        ax.set(xlabel='Scalp Coupling Index', ylabel='Count', xlim=[0, 1])
        raw_od.info
    raw_od.info['bads'] = list(compress(raw_od.ch_names, sci < 0.5))
    return raw_od

def GetRawHaemo(raw_od, start = 2000, duration = 500, plot = False):
    ### band filter
    filt_ODs = raw_od.filter(0.01, 0.1, h_trans_bandwidth=0.03, l_trans_bandwidth=0.005)  # ~0.1 ## Gilbert Changed here. 
    ### MBLL - Conversion from ODs to hemoglobin concentration changes
    raw_haemo = mne.preprocessing.nirs.beer_lambert_law(filt_ODs, ppf=6)
    ### plot 
    if plot:
        raw_haemo.plot(n_channels=len(raw_haemo),start=start, ## 1-40 hbo 41-80 hbr raw_haemo.ch_names[:40]
        duration=duration, show_scrollbars=True,
            scalings='auto', clipping=None,)
    return raw_haemo

### get events 
def RecorverEventsFromCSV(df_raw_B,df_raw_P,df_raw_G,raw_haemo,subject, plot = False):
    events, event_dict = mne.events_from_annotations(raw_haemo)
    events_clean = RemoveDuplicateEvents(events)
    ### covert utc to unix 
    # unixtime = time.mktime(raw_haemo.info["meas_date"].timetuple())
    unixtime = time.mktime(raw_haemo.info["meas_date"].timetuple())

    # s33
    ### subject
    s = subject

    df_group_B = GetSubjectEvents(s,df_raw_B,unixtime)
    df_group_P = GetSubjectEvents(s,df_raw_P,unixtime)
    df_group_G = GetSubjectEvents(s,df_raw_G,unixtime)

    ### get event data 
    Con_B,Anc_B,Asy_B,Syn_B = GetConditions(df_group_B)
    Con_P,Anc_P,Asy_P,Syn_P = GetConditions(df_group_P)
    Con_G,Anc_G,Asy_G,Syn_G = GetConditions(df_group_G)

    ### covert events 

    ### Cube B
    arr_Anc_S_B = AppendEventsToArr(Anc_B['StartFrame'],3)
    arr_Anc_E_B = AppendEventsToArr(Anc_B['EndCountFrame'],4)

    arr_Con_S_B = AppendEventsToArr(Con_B['StartFrame'],1)
    arr_Con_E_B = AppendEventsToArr(Con_B['EndCountFrame'],2)

    arr_Asy_S_B = AppendEventsToArr(Asy_B['StartFrame'],5)
    arr_Asy_E_B = AppendEventsToArr(Asy_B['EndCountFrame'],6)

    arr_Syn_B_S_B = AppendEventsToArr(Syn_B['StartFrame'],7)
    arr_Syn_B_E_B = AppendEventsToArr(Syn_B['EndCountFrame'],8)

    ### Cube P
    arr_Con_S_P = AppendEventsToArr(Con_P['StartFrame'],1)
    arr_Con_E_P = AppendEventsToArr(Con_P['EndCountFrame'],2)

    arr_Anc_S_P = AppendEventsToArr(Anc_P['StartFrame'],3)
    arr_Anc_E_P = AppendEventsToArr(Anc_P['EndCountFrame'],4)

    arr_Asy_S_P = AppendEventsToArr(Asy_P['StartFrame'],5)
    arr_Asy_E_P = AppendEventsToArr(Asy_P['EndCountFrame'],6)

    arr_Syn_B_S_P = AppendEventsToArr(Syn_P['StartFrame'],7)
    arr_Syn_B_E_P = AppendEventsToArr(Syn_P['EndCountFrame'],8)


    ### Cube G
    arr_Con_S_G = AppendEventsToArr(Con_G['StartFrame'],1)
    arr_Con_E_G = AppendEventsToArr(Con_G['EndCountFrame'],2)

    arr_Anc_S_G = AppendEventsToArr(Anc_G['StartFrame'],3)
    arr_Anc_E_G = AppendEventsToArr(Anc_G['EndCountFrame'],4)

    arr_Asy_S_G = AppendEventsToArr(Asy_G['StartFrame'],5)
    arr_Asy_E_G = AppendEventsToArr(Asy_G['EndCountFrame'],6)

    arr_Syn_B_S_G = AppendEventsToArr(Syn_G['StartFrame'],7)
    arr_Syn_B_E_G = AppendEventsToArr(Syn_G['EndCountFrame'],8)

    
    arr = np.concatenate((
        arr_Anc_S_B,arr_Anc_E_B,arr_Con_S_B,arr_Con_E_B,arr_Asy_S_B,arr_Asy_E_B,arr_Syn_B_S_B,arr_Syn_B_E_B,
        arr_Anc_S_P,arr_Anc_E_P,arr_Con_S_P,arr_Con_E_P,arr_Asy_S_P,arr_Asy_E_P,arr_Syn_B_S_P,arr_Syn_B_E_P,
        arr_Anc_S_G,arr_Anc_E_G,arr_Con_S_G,arr_Con_E_G,arr_Asy_S_G,arr_Asy_E_G,arr_Syn_B_S_G,arr_Syn_B_E_G,

    ))



    # Dict = {'whiteS': 1,'whiteE': 2,'PurpleS': 3,'PurpleE': 4,'GreenS': 5,'GreenE': 6,'BlueS': 7,'BlueE': 8}
    Dict = {'Control_Start': 1,'Control_End': 2,'Anc_Start': 3,'Anc_End': 4,'Asy_Start': 5,'Asy_End': 6,'Syn_Start': 7,'Syn_End': 8,}
    # print(events_clean)
    event_id = Dict

    if plot:
        fig = mne.viz.plot_events(arr, event_id=Dict,
                            sfreq=raw_haemo.info['sfreq'])
        fig.subplots_adjust(right=0.7)  # make room for the legend
    
    eventArr = arr
    eventDict = Dict
    return eventArr,eventDict

### get subject data from total 
def GetSubjectEvents(s,df,unixtime):
    df_group = df.groupby('Subject').get_group(s).reset_index()
    df_group['StartCount'] = df_group['Start']-unixtime
    df_group['EndCount'] = df_group['End']-unixtime
    ### seconds to frame 
    df_group['StartFrame'] =(df_group['StartCount']*5.1).astype('int32')
    df_group['EndCountFrame'] = (df_group['EndCount']*5.1).astype('int32')

    return df_group

def RemoveDuplicateEvents(events):
    new_array = [tuple(row) for row in events]
    uniques = np.unique(new_array,axis=0)
    return uniques

### list is faster than np.array
def AppendEventsToArr(s,id):
    # arr =  np.array((0,3))
    arr = []
    for index, value in s.items():
        # print(f"Index : {index}, Value : {value}")
        # add = np.array([value,0,id])
        arr.append([value,0,id])
    l = np.asarray(arr)
    # print("----------Event Arr-------------")
    # print(l)
    return l

def GetConditions(df):
    Con = df.loc[df['Condition'] == 'control'].reset_index()
    Anc = df.loc[df['Condition'] == 'Anchoring'].reset_index()
    Asy = df.loc[df['Condition'] == 'Asynchronized'].reset_index()
    Syn = df.loc[df['Condition'] == 'Synchronized'].reset_index()
    return Con,Anc,Asy,Syn

### df data analysis 

def TrackingNoiseTrial(df, threshold = 1,):
    df_out = pd.DataFrame()
    df_out['pandas'] = df.rolling(60).mean().fillna(method='bfill').fillna(method='ffill')

    difference = np.abs(df - df_out['pandas'])

    outlier_idx = difference > threshold
    noise = len(df[outlier_idx])
    # print("Sum of Noise:", noise)
    if(noise > 10):
        df.loc[:] = None

    return df

def DrawNoiseTrial(df, threshold = 1,):
    df['pandas'] = df['raw'].rolling(60).mean().fillna(method='bfill').fillna(method='ffill')

    difference = np.abs(df['raw'] - df['pandas'])

    outlier_idx = difference > threshold
    print("Sum of Noise:", len(df['raw'][outlier_idx]))

    figsize = (7, 2.75)
    kw = dict(marker='o', linestyle='none', color='r', alpha=0.3)
    fig, ax = plt.subplots(figsize=figsize)
    df['raw'].plot()
    df['raw'][outlier_idx].plot(**kw)
    _ = ax.set_ylim(-2, 2)

### General function for spike data detection 
def GetCleanDataFrameFromEpochFrame(df_epochs,conditions = ['condition','epoch'],ChannelIndex = 3):
    ### df_epochs: converted from mne epochs data  epochs.to_data_frame(picks=None) 
    ### df_clean: outputs
    ### trials: grouped epoch data
    trials = df_epochs.groupby(conditions,as_index=False)
    df_clean = pd.DataFrame()
    for index, group in trials:
        df_g = group.iloc[:,ChannelIndex:].apply(lambda x: TrackingNoiseTrial(x))
        df_new = pd.concat([group.iloc[:,0:ChannelIndex],df_g],axis=1)
        df_clean = pd.concat([df_clean,df_new],axis=0)
        
    return df_clean
```

```python
def readValidData(filename):
    try:
        intensity, subject = GetRawIntensity(fileName=filename)
        if subject not in ['14', 'Subject_14:_2nd_round','17','18','18:_trial_8_-_10', '23', '28']:
            if subject[0] in ['s', 'S']:
                subject = eval(subject[8:10])
            elif subject == '22_round2':
                subject = 22
            else:
                subject = eval(subject)
            
            return (intensity, subject)
    except RuntimeError:
        print("error....")
    return False

def GetIndividual(directory, plot = False):
    ### get indiviual data 
    data = readValidData(directory)

    if data:
        raw_intensity = data[0]
        subject = data[1]
        del data
        print(subject)
        if subject in df_raw_B.Subject:
            raw_od = GetRawOd(raw_intensity, plot = plot)
            raw_haemo = GetRawHaemo(raw_od, plot = plot)
            
            ### get events
            eventArr,eventDict = RecorverEventsFromCSV(df_raw_B,df_raw_P,df_raw_G,raw_haemo,subject)
            ### get qualified block signal according to event
            reject_criteria = dict(hbo=80e-6)
            tmin, tmax = -10, 30 #  time duration 

            epochs_All = mne.Epochs(raw_haemo, eventArr, event_id=eventDict,
                                tmin=tmin, tmax=tmax,event_repeated="merge",
                                reject=reject_criteria, reject_by_annotation=True,
                                proj=True, baseline=(None, 0), preload=True,
                                detrend=None, verbose=True)# ssp projection vectors; begining baseline correction; 
            # epochs_All.plot_drop_log()
            df = epochs_All.to_data_frame(picks=None)
            del epochs_All
            df['pid'] = subject
            return df
    return 'invalid'
```

```python
def GetCorrelationEpoch(df, epochid, cor_lst):
    df_selected = df.query("epoch == @epochid")
    corr_cof = df_selected[cor_lst].corr()

    corr_cof = 0.5* np.log(1 + corr_cof)/(1 - corr_cof) # see https://www.nature.com/articles/s41598-018-33439-2: Fisher z-transformation
    return corr_cof


def GetCorrelationCondition(df,condition, cor_lst):
    dataSelected = df.query("condition == @condition")
    metaCor = GetCorrelationEpoch(dataSelected, dataSelected.epoch.unique()[0], cor_lst)
    num = 1
    for epochid in dataSelected.epoch.unique()[1:]:
        metaCor += GetCorrelationEpoch(dataSelected, epochid, cor_lst)
        num +=1

    return metaCor/num

def GetCorrelationPerson(df, pid, cor_lst):
    data_selected = df.query("pid == @pid")
    result = pd.DataFrame(columns=['ID',  'condition','correlation'])
    for cond in data_selected.condition.unique():
        condition_cor = pd.DataFrame([{'ID':pid, 'condition':cond, 'correlation':GetCorrelationCondition(data_selected, cond, cor_lst)}])

        result = pd.concat([result, condition_cor], ignore_index=True)
    return result

def GetCorrelation(df, cor_lst):
    result = pd.DataFrame(columns=['ID',  'condition','correlation'])
    for pid in df.pid.unique():
        pid_cor = GetCorrelationPerson(df, pid, cor_lst)
        result = pd.concat([result, pid_cor], ignore_index=True)
    return result
```

# get Connection ratio and connection strength
The number of significant connections of a network was calculated by counting all connections belonging to the network that had an absolute z-value greater than the threshold. In this study, the threshold range was chosen from 0.2 to 0.7. The connection ratio (CR) was then computed as a ratio of the significant connections on the network total connections. The connection strength (CS) of a network was the average of the absolute z-values of all connections.

```python
def GetCR(df):
    validdf = df.copy()
    validdf.replace([np.inf, -np.inf], 0, inplace=True)
    CR = ((validdf.abs() > cr_threhold).sum().sum())/(validdf.shape[0]*(validdf.shape[1]-1))
    return CR

def GetCS(df):
    validdf = df.copy()
    validdf.replace([np.inf, -np.inf], 0, inplace=True)
    return (validdf.abs().sum().sum())/(validdf.shape[0] * (validdf.shape[1] - 1))

def GetCRExt(df):
    validdf = df.copy()
    validdf.replace([np.inf, -np.inf], 0, inplace=True)
    CR = ((validdf.abs() > cr_threhold).sum().sum())/(validdf.shape[0]*(validdf.shape[1]))
    return CR

def GetCSExt(df):
    validdf = df.copy()
    validdf.replace([np.inf, -np.inf], 0, inplace=True)
    return validdf.abs().sum().sum()/(validdf.shape[0] * (validdf.shape[1]))

cr_threhold = 0.3 # ref: 0.2-0.7
full_connect['CR'] = full_connect['correlation'].apply(GetCR)
full_connect['CS'] = full_connect['correlation'].apply(GetCS)
```

```python
def GetPerParcellation(data, collst1, collst2):
    df = data.copy()
    result =  pd.DataFrame(df.loc[collst1[0],collst2])
    for col in collst1[1:]:
        new_col = pd.DataFrame(df.loc[col,collst2])
        result = pd.concat([result, new_col], axis= 1)
    return result

def InternalParcellationStat(parcel):
    cr = GetCR(parcel)
    cs = GetCS(parcel)
    return cr, cs

def ExternalParcellationStat(parcel):
    cr = GetCRExt(parcel)
    cs = GetCSExt(parcel)
    return cr, cs

def GetGroupParcellation(data, collst1, collst2, featureName, internal):
    df = data.copy()
    df[featureName + '_CR'] = 0
    df[featureName + '_CS'] = 0
    for index, row in df.iterrows():
        if internal:
            cr, cs = InternalParcellationStat(GetPerParcellation(row['correlation'], collst1, collst2))
        else:
            cr, cs = ExternalParcellationStat(GetPerParcellation(row['correlation'], collst1, collst2))
        df.loc[index, featureName + '_CR'] = cr
        df.loc[index, featureName + '_CS'] = cs
    return df

def GetAllInternalParacellation(data, gpDict):
    df = data.copy()

    for gpkey in gpDict.keys():
        df = GetGroupParcellation(df, gpDict[gpkey], gpDict[gpkey], gpkey, internal=True)
    return df

def GetAllExternalParacellation(data, gpDict):
    df = data.copy()

    # make combinations of external parracellation
    combination = []
    for i in range(len(gpDict)):
        first = list(gpDict.keys())[i]
        for j in range(len(gpDict) - i - 1):
            second = list(gpDict.keys())[i+j+1]
            combination.append([first, second])

    for i in combination:
        df = GetGroupParcellation(df, gpDict[i[0]], gpDict[i[1]], i[0]+'_'+i[1],  internal=False)
    return df
```

```python

def Statics(condi,col, data):

    temp2 = data.copy()
    
    
    sig = []
    
    for i in range(len(condi)):
        for j in range(i+1, len(condi)):
            res = wilcoxon(temp2.loc[temp2.condition == condi[i], col], 
                   temp2.loc[temp2.condition == condi[j], col])
            print(condi[i], ' V.S. ', condi[j],': ', res)
            sig.append(res[1])
            
    bars = np.arange(len(sig))
    
    return sig, bars


def barplot_annotate_brackets(num1, num2, data, center, height, yerr=None, dh=.05, barh=.05, fs=None, maxasterix=None, textoffset = 0.):
    """ 
    Annotate barplot with p-values.

    :param num1: number of left bar to put bracket over
    :param num2: number of right bar to put bracket over
    :param data: string to write or number for generating asterixes
    :param center: centers of all bars (like plt.bar() input)
    :param height: heights of all bars (like plt.bar() input)
    :param yerr: yerrs of all bars (like plt.bar() input)
    :param dh: height offset over bar / bar + yerr in axes coordinates (0 to 1)
    :param barh: bar height in axes coordinates (0 to 1)
    :param fs: font size
    :param maxasterix: maximum number of asterixes to write (for very small p-values)
    """
    
    if type(data) is str:
        text = data
    else:
        # * is p < 0.05
        # ** is p < 0.005
        # *** is p < 0.0005
        # etc.
        text = ''
        p = .05

        while data < p:
            text += '*'
            p /= 10.

            if maxasterix and len(text) == maxasterix:
                break

        if len(text) == 0:
            text = 'n. s.'

    lx, ly = center[num1], height
    rx, ry = center[num2], height

    if yerr:
        ly += yerr[num1]
        ry += yerr[num2]

    ax_y0, ax_y1 = plt.gca().get_ylim()
    dh *= (ax_y1 - ax_y0)
    barh *= (ax_y1 - ax_y0)

    y = max(ly, ry) + dh

    barx = [lx, lx, rx, rx]
    bary = [y, y+barh, y+barh, y]
    mid = ((lx+rx)/2, y+barh+textoffset)

    plt.plot(barx, bary, c='black')

    kwargs = dict(ha='center', va='bottom')
    if fs is not None:
        kwargs['fontsize'] = fs

    plt.text(*mid, text, **kwargs)
```

```python
def StaticsMatrics(condi,col, data):

    temp2 = data.copy()
    
    sig = []
    condi_pairs = []
    for i in range(len(condi)):
        for j in range(i+1, len(condi)):
            res = wilcoxon(temp2.loc[temp2.condition == condi[i], col], 
                   temp2.loc[temp2.condition == condi[j], col])
            condi_pairs.append(condi[i] + ' V.S. ' + condi[j])
            # print(condi[i], ' V.S. ', condi[j],': ', res)
            sig.append(res[1])
            
    return sig, condi_pairs
```

```python
def barplot_annotate_brackets_subplot(num1, num2, data, center, height, axis, yerr=None, dh=.05, barh=.05, fs=None, maxasterix=None, textoffset = 0.):
    """ 
    Annotate barplot with p-values.

    :param num1: number of left bar to put bracket over
    :param num2: number of right bar to put bracket over
    :param data: string to write or number for generating asterixes
    :param center: centers of all bars (like plt.bar() input)
    :param height: heights of all bars (like plt.bar() input)
    :param yerr: yerrs of all bars (like plt.bar() input)
    :param dh: height offset over bar / bar + yerr in axes coordinates (0 to 1)
    :param barh: bar height in axes coordinates (0 to 1)
    :param fs: font size
    :param maxasterix: maximum number of asterixes to write (for very small p-values)
    """
    
    if type(data) is str:
        text = data
    else:
        # * is p < 0.05
        # ** is p < 0.005
        # *** is p < 0.0005
        # etc.
        text = ''
        p = .05
        cnt = 0
        while data < p and cnt < 3:
            text += '*'
            p /= 10.
            cnt +=1
            if maxasterix and len(text) == maxasterix:
                break

        if len(text) == 0:
            text = 'n. s.'

    lx, ly = center[num1], height
    rx, ry = center[num2], height

    if yerr:
        ly += yerr[num1]
        ry += yerr[num2]

    ax_y0, ax_y1 = axis.get_ylim()
    dh *= (ax_y1 - ax_y0)
    barh *= (ax_y1 - ax_y0)

    y = max(ly, ry) + dh

    barx = [lx, lx, rx, rx]
    bary = [y, y+barh, y+barh, y]
    mid = ((lx+rx)/2, y+barh+textoffset)

    axis.plot(barx, bary, c='black')

    kwargs = dict(ha='center', va='bottom')
    if fs is not None:
        kwargs['fontsize'] = fs

    axis.text(*mid, text, **kwargs)
```

```python
### get arr from each condition and subject 
def rows_generator(df):
    i = 0
    while (i+40) <= df.shape[0]:
        yield df.iloc[i:(i+40):1, :]
        i += 40


def GetCorrelationArr(condition,ID,df,fre,window):
    # df_selected = df.query("ID == @ID").query("Condition == @condition") 
    windowsize =int(window*fre)
    # selected_corr = df_selected[['LDLPFC','MDLPFC','RDLPFC','LPM','SMA','RPM',]]
    corr_cof = df.rolling(windowsize).corr(pairwise=True)
    i = 1
    cof_arr = []
    ### convert dataframe to matrix array  
    for df in rows_generator(corr_cof):
        # print(f'Corr #{i}')
        # print(df.to_numpy())
        a = df.to_numpy()
        cof_arr.append(a)
        i += 1
        
    print('-----------',condition,'------------')
    # print(cof_arr[-1])
    ### resample  xs[1::4] start, stop, step
    print(len(cof_arr))
    newArr = cof_arr[449::fre]
    print(len(newArr))
    print(newArr[-1])
    return newArr

# modify here Gilbert
def GetCorrelationEpoch(df, pid,condition, epochid, cor_lst):
    df_selected = df.query("pid == @pid").query("Condition == @condition").query("epoch == @epochid")
    # selected_corr = df_selected[['LDLPFC','MDLPFC','RDLPFC','LPM','SMA','RPM',]]
    corr_cof = df_selected[cor_lst].corr(pairwise=True)
    # i = 1
    # cof_arr = []
    # ### convert dataframe to matrix array  
    # for df in rows_generator(corr_cof):
    #     # print(f'Corr #{i}')
    #     # print(df.to_numpy())
    #     a = df.to_numpy()
    #     cof_arr.append(a)
    #     i += 1

    return corr_cof

def GetAllCoefs(f,window,df_input,max_id_index):
    ### output df
    df_coef =  pd.DataFrame(columns=['ID',  'Control','Landmark','Survey','Route'])
    for ID in range(1,max_id_index):
        ### get coefs 
        coef_control = GetCorrelationArr('Control',ID,df_input,f,window)
        coef_route = GetCorrelationArr('Route_Task',ID,df_input,f,window)
        coef_survey = GetCorrelationArr('Survey_Task',ID,df_input,f,window)
        coef_landmark = GetCorrelationArr('Landmark_Task',ID,df_input,f,window)
        ### loop 
        len_coef = len(coef_control)
        print("arr length:", len_coef)
        for i in range(0,len_coef):
            df_new= {'ID': i, 'Control': coef_control[i], 'Route': coef_route[i],
            'Survey': coef_survey[i], 'Landmark': coef_landmark[i]
            }
            df_coef = df_coef.append(df_new, ignore_index = True)
        # conc_arr.append(coef_control)

        conc_arr = np.concatenate([coef_control, coef_route,coef_survey,coef_landmark], axis=0)
        print("arr shape:", conc_arr.shape)
        if ID == 1:
            arr_all =conc_arr
        else:
            arr_all= np.concatenate([arr_all,conc_arr], axis=0)
        ### 
    print("df shape:", df_coef.shape)
    # conc_arr = np.array(conc_arr)
    print("arr shape:", arr_all.shape)
    return df_coef,arr_all
```
